```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier  # Import Decision Tree Classifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

ùêãùê®ùêöùêùùê¢ùêßùê† ùêÉùêöùê≠ùêö ùêöùêßùêù ùêíùêûùê•ùêûùêúùê≠ùê¢ùêßùê† ùêÖùêûùêöùê≠ùêÆùê´ùêûùê¨

```python
# Load the dataset
data = pd.read_csv(r"C:\Users\trish\Desktop\ML4PS-master\onlinefoods.csv")

# Convert non-numeric columns to numeric using LabelEncoder
label_encoders = {}
for column in data.columns:
    if data[column].dtype == 'object':
        label_encoders[column] = LabelEncoder()
        data[column] = label_encoders[column].fit_transform(data[column])

# Selecting features and target
selected_features = ['Age', 'Family size']
X = data[selected_features]  # Using only selected features
y = data['Gender']
```

ùêíùê©ùê•ùê¢ùê≠ùê≠ùê¢ùêßùê† ùê≠ùê°ùêû ùêùùêöùê≠ùêö ùê¢ùêßùê≠ùê® ùêìùê´ùêöùê¢ùêßùê¢ùêßùê† ùêöùêßùêù ùêìùêûùê¨ùê≠ùê¢ùêßùê† ùê¨ùêûùê≠ùê¨ ùêöùêßùêù ùêàùêßùê¢ùê≠ùê¢ùêöùê•ùê¢ùê≥ùêû ùê¶ùê®ùêùùêûùê•.

```python
# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Initialize Decision Tree Classifier model with max_depth=1 to make it very simple
clf = DecisionTreeClassifier(max_depth=1)
```

ùêåùê®ùêùùêûùê• ùêìùê´ùêöùê¢ùêßùê¢ùêßùê† ùêöùêßùêù ùêÑùêØùêöùê•ùêÆùêöùê≠ùê¢ùê®ùêß

```python
# Train the classifier
clf.fit(X_train, y_train)

# Predictions on training and testing sets
train_preds = clf.predict(X_train)
test_preds = clf.predict(X_test)

# Calculate accuracy on training and testing sets
train_accuracy = accuracy_score(y_train, train_preds)
test_accuracy = accuracy_score(y_test, test_preds)
```

ùêëùêûùê¨ùêÆùê•ùê≠

```python

# Print the overall accuracy of training and testing sets
print("Overall Training Accuracy:", train_accuracy)
print("Overall Testing Accuracy:", test_accuracy)

# Data
labels = ['Training Accuracy', 'Testing Accuracy']
accuracies = [train_accuracy, test_accuracy]

# Plot
plt.figure(figsize=(8, 6))
plt.plot(labels, accuracies, marker='o', linestyle='-')
plt.title('Decision Tree Model Accuracy')
plt.ylabel('Accuracy')
plt.ylim(0, 1)  # Limit y-axis from 0 to 1 for better visualization
plt.grid(True)  # Add grid lines
plt.show()
```
